```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the Iris dataset
iris = sns.load_dataset("iris")
```

```python
# 1. Pairplot of the Dataset
sns.pairplot(iris, hue="species", markers=["o", "s", "D"])
plt.suptitle("Pairplot of Iris Dataset", y=1.02)
plt.show()
```

```python
# 2. Distribution of Each Feature
plt.figure(figsize=(12, 8))
for i, feature in enumerate(iris.columns[:-1]):
    plt.subplot(2, 2, i + 1)
    sns.histplot(iris[feature], kde=True, bins=20)
    plt.title(f"Distribution of {feature}")
plt.tight_layout()
plt.show()
```

```python
# 3. Boxplot of Each Feature by Species
plt.figure(figsize=(12, 8))
for i, feature in enumerate(iris.columns[:-1]):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(x="species", y=feature, data=iris)
    plt.title(f"Boxplot of {feature} by Species")
plt.tight_layout()
plt.show()
```

```python
# 4. Heatmap of Correlation Matrix
plt.figure(figsize=(8, 6))
# Exclude the 'species' column
numeric_features = iris.drop(columns=["species"])
corr = numeric_features.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Heatmap of Correlation Matrix")
plt.show()
```
